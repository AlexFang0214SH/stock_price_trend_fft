```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataG = pd.read_csv('Google_Stock_Price_Train.csv', thousands=',')
```

```python
dataG.shape
```

```python
dataG.head()
```

```python
data = dataG[['Date', 'Close']]
data['Close'] = pd.to_numeric(data['Close'])
```

```python
data.head(n=10)
```

```python
data.plot(y='Close',x='Date', figsize=(15, 7))
plt.show()
```

```python
close_fft = np.fft.fft(np.asarray(data['Close'].tolist()))
```

```python
fft_df = pd.DataFrame({'fft':close_fft})
fft_df['abs'] = fft_df['fft'].apply(lambda x: np.absolute(x))
fft_df['angle'] = fft_df['fft'].apply(lambda x: np.angle(x))
#fft_df.sort_values(by=['abs'], ascending=False, inplace=True)
#fft_df.reset_index(inplace=True)
#fft_df.drop(columns=['index'], inplace=True)
```

```python
fft_df.head(n=20)
```

```python
# plot all components
plt.figure(figsize=(15, 7))
plt.plot(np.fft.ifft(np.asarray(fft_df['fft'].tolist())))
plt.show()
```

```python
# plot first 100 components

fft_list = np.asarray(fft_df['fft'].tolist())
fft_list[100:] = 0
ifft_list = np.fft.ifft(fft_list)
plt.figure(figsize=(15, 7))
plt.plot(ifft_list)
plt.show()
```

```python
# plot first 10 components

fft_list = np.asarray(fft_df['fft'].tolist())
fft_list[10:] = 0
ifft_list = np.fft.ifft(fft_list)
plt.figure(figsize=(15, 7))
plt.plot(ifft_list)
plt.show()
```
