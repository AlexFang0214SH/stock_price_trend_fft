```python
data = dataG[['Date', 'Close']]
```

```python
data.head(n=17)
```

```python
from scipy.fftpack import fft
```

```python
from scipy.fftpack import ifft
```

```python
data['Close'] = data['Close'].apply(lambda x: x.replace(',', ''))
```

```python
data['Close'] = pd.to_numeric(data['Close'])
```

```python
data['FFT'] = fft(data['Close'].tolist())
```

```python
data['iFFT'] = ifft(data['Close'].tolist())
```

```python
data['FFT_r'] = data['FFT'].apply(lambda x: x.real)
```

```python
data['FFT_i'] = data['FFT'].apply(lambda x: x.imag)
```

```python
data['iFFT_r'] = data['iFFT'].apply(lambda x: x.real)
```

```python
data['iFFT_i'] = data['iFFT'].apply(lambda x: x.imag)
```

```python
data.head(n=17)
```

```python
data['amp'] = data.apply(lambda row: amp_fn(row['FFT_r'], row['FFT_i']), axis=1)
```

```python
data['angle'] = data.apply(lambda row: angle_fn(row['FFT_r'], row['FFT_i']), axis=1)
```

```python
data['i_amp'] = data.apply(lambda row: amp_fn(row['iFFT_r'], row['iFFT_i']), axis=1)
```

```python
data['i_angle'] = data.apply(lambda row: angle_fn(row['iFFT_r'], row['iFFT_i']), axis=1)
```

```python
def amp_fn(x, y):
    return np.sqrt(x**2 + y**2)
```

```python
def angle_fn(x, y):
    if y==0:
        y=x
    return np.arctan(x / y)
```
